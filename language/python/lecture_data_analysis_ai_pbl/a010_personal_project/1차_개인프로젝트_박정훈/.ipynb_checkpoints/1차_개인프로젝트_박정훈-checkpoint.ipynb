```python
"""
목적 : 2022 년의 각 시군구별 한 가구당 평균 인구수를 구하고 지도에 시각화 해보자
인구수 데이터와 가구수 데이터를 사용한다
"""
```

```python
import pandas as pd
```

```python
# 성별 인구 수 가공
population = pd.read_csv("./01_수집_데이터/행정구역별성별인구수_2022.csv")
population
```

```python
"""
population.rename(
    columns = {
        "행정구역(시군구)별(1)" : "광역시도",
        "행정구역(시군구)별(2)" : "시군구",
        "2022" : "총인구수",
        "2022.1" : "남자인구수",
        "2022.2" : "여자인구수",
    },
    inplace = True,
)
"""
population.columns = [ "광역시도", "시군구", "총인구수", "남자인구수", "여자인구수" ]
population.drop([0], inplace = True)
population.reset_index(drop = True, inplace = True)
population
```

```python
# 초기 데이터의 0번째 row로 인해 object type인 column을 int type으로 형변환
population = population.astype({
    "광역시도" : object,
    "시군구" : object,
    "총인구수" : int,
    "남자인구수" : int,
    "여자인구수" : int
})
population.info()
```

```python
# 불필요한 행 삭제
rowsToRemain = (population["시군구"] != "소계")
population = population[rowsToRemain]
population.reset_index(drop = True, inplace = True)
population
```

```python
# 고유한 열 생성
id = [population.loc[index, "광역시도"] + population.loc[index, "시군구"] for index in population.index]
# population["id"] = id
population.insert(
    loc = 0,
    column = "id",
    value = id
)
population
```

```python
# 중복 데이터 체크
population["id"].duplicated().value_counts()
```

```python
# 주민등록 세대 수 가공
family = pd.read_csv("./01_수집_데이터/행정구역별주민등록세대수_2022.csv")
family.info
```

```python
family.columns = [ "광역시도", "시군구", "가구수" ]
family
```

```python
rowsToRemain = (family["시군구"] != "소계")
family = family[rowsToRemain]
family.reset_index(drop = True, inplace = True)
family
```

```python
id = [family.loc[index, "광역시도"] + family.loc[index, "시군구"] for index in family.index]
family.insert(
    loc = 0,
    column = "id",
    value = id
)
family
```

```python
# 중복 데이터 체크
family["id"].duplicated().value_counts()
```

```python
# population과 family의 id 일치 여부 체크
if set(population["id"]) == set(family["id"]) :
    print("same")
else :
    print("not same")
```

```python
# 두 테이블 합치기
totalTable = pd.merge(
    left = population,
    right = family,
    how = "outer",
    on = ["id", "광역시도", "시군구"],
)
totalTable
```

```python
# 결측치 검사
totalTable.info()
```

```python
# 총인구수 이상치 검사
totalTable["총인구수"].describe()
```

```python
# 0 값 제거
notZero = (totalTable["총인구수"] != 0)
totalTable = totalTable[notZero]
totalTable.reset_index(drop = True, inplace = True)
totalTable
```

```python
# 가구수 이상치 검사
totalTable["가구수"].describe()
```

```python
# 전처리 완료 데이터 저장
totalTable.to_csv("./02_전처리완료_데이터/total_table.csv")
```

```python
# 평균 가구원수 열 생성
averageFamilyMember = totalTable["총인구수"] / totalTable["가구수"]
totalTable.insert(
    loc = len(totalTable.columns),
    column = "평균가구원수",
    value = averageFamilyMember,
)
totalTable
```

```python
totalTable["평균가구원수"].describe()
```

```python
# 시각화
# 평균가구원수 상위 10개와 하위 10개를 막대 그래프로 그려보자
import matplotlib.pyplot as plt

plt.rc("font", family = "AppleGothic")
```

```python
top10 = totalTable.sort_values(
    "평균가구원수",
    ascending = False,
).head(10)
top10
```

```python
bottom10 = totalTable.sort_values(
    "평균가구원수",
    ascending = False,
).tail(10)
bottom10
```

```python
plt.figure(figsize = (10, 5))
plt.title("평균가구원수 상위 10", fontsize = "17")

# plt.bar(top10["id"], top10["평균가구원수"], color="lightgrey")

for index in top10.index :
    x = top10.loc[index, "id"]
    y = top10.loc[index, "평균가구원수"]
    plt.bar(x, y, color="lightgrey")
    plt.text(x, y, "%.2f"%y, horizontalalignment="center", verticalalignment="bottom")

plt.xlabel("지역")
plt.ylabel("평균 가구원 수")
plt.xticks(rotation = 30)

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].set_title("평균가구원수 상위 10")

for index in top10.index :
    x = top10.loc[index, "id"]
    y = top10.loc[index, "평균가구원수"]
    axs[0].bar(x, y, color="lightgrey")
    axs[0].text(x, y, "%.2f"%y, horizontalalignment="center", verticalalignment="bottom")

axs[0].set_xlabel("지역")
axs[0].set_ylabel("평균 가구원 수")
axs[0].tick_params(labelrotation=45)

axs[1].set_title("평균가구원수 하위 10")

for index in bottom10.index :
    x = bottom10.loc[index, "id"]
    y = bottom10.loc[index, "평균가구원수"]
    axs[1].bar(x, y, color="lightgrey")
    axs[1].text(x, y, "%.2f"%y, horizontalalignment="center", verticalalignment="bottom")

axs[1].set_xlabel("지역")
axs[1].set_ylabel("평균 가구원 수")
axs[1].tick_params(labelrotation=45)

plt.show()
```
